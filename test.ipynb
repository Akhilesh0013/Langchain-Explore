# Web Based Loader

```python
from langchain_community.document_loaders import WebBaseLoader
import bs4
```

```python
web_loader = WebBaseLoader( web_path= ['https://www.universetoday.com/articles/the-impossible-black-holes-that-shouldnt-exist'] ,
                           bs_kwargs= dict(parse_only = bs4.SoupStrainer(class_ = ("article-header" , "article-content")) )
                                                      
                    )
```

```python
web_loader.load()
```

```python
docs = web_loader.load()
```

```python
docs
```

# Text Splitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
r_splitter = RecursiveCharacterTextSplitter( chunk_size = 200 , chunk_overlap = 50 )
```

```python
final_docs = r_splitter.split_documents(docs)
```

```python
final_docs
```
